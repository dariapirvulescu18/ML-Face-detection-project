```python
from ultralytics import YOLO
import cv2
import numpy as np
from pathlib import Path
import os
```

```python
##########################################
# schimbare fisier
test_images_dir = Path("./testare/")
#########################################
os.makedirs("./351_Pirvulescu_Daria",exist_ok=True)
os.makedirs("./351_Pirvulescu_Daria/task1_bonus",exist_ok=True)
os.makedirs("./351_Pirvulescu_Daria/task2_bonus",exist_ok=True)
```

```python
model = YOLO("YOLO_toate_fetele/weights/best.pt")  
detectii_finale = []
scoruri_finale = []
file_names = []
```

```python
def detecteaza_fete(img_path, model):
    print(img_path)
    img = cv2.imread(str(img_path))
    results = model.predict(source=img, save=False, verbose=False)  
    detectii = results[0].boxes.xyxy.cpu().numpy().astype(int)  
    scoruri = results[0].boxes.conf.cpu().numpy()  
    detectii_finale.append(detectii)
    scoruri_finale.extend(scoruri)  
    file_names.extend([img_path.name] * len(scoruri))  
```

```python
for image_file in test_images_dir.glob("*.jpg"):  
    detecteaza_fete(image_file, model)

all_detections = np.vstack(detectii_finale) if detectii_finale else np.array([])

np.save("./351_Pirvulescu_Daria/task1_bonus/detections_all_faces.npy", all_detections)
np.save("./351_Pirvulescu_Daria/task1_bonus/scores_all_faces.npy", np.array(scoruri_finale))
np.save("./351_Pirvulescu_Daria/task1_bonus/file_names_all_faces.npy", np.array(file_names))
```

```python
files= os.listdir("./testare/")
os.makedirs("./detected_faces_YOLO/",exist_ok=True)
for file in files:
    img = cv2.imread("./testare/"+ file)
    for det, score , name in zip(all_detections, scoruri_finale, file_names):
        if name== file:
            x_min, y_min, x_max, y_max = det
            score = (score - min(scoruri_finale)) / (max(scoruri_finale) - min(scoruri_finale) + 1e-6)
            red_intensity = int(min(max(score * 255, 0), 255))
            cv2.rectangle(img, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0,0,red_intensity), 2)
    cv2.imwrite(f"./detected_faces_YOLO/{file}", img)
```

```python
model = YOLO("./YOLO_personaje/weights/best.pt")  

detectii_dict = { "dad": [], "deedee": [], "mom": [], "dexter": [] }
scoruri_dict = { "dad": [], "deedee": [], "mom": [], "dexter": [] }
file_names_dict = { "dad": [], "deedee": [], "mom": [], "dexter": [] }

class_map = {
    0: "dad",      
    1: "deedee",
    2: "dexter",    
    3: "mom",       
}
```

```python
def recunoastere_personaje(img_path, model):
    img = cv2.imread(str(img_path))
    print(img_path)
    results = model.predict(source=img, save=False, verbose=False) 

    detections = results[0].boxes.xyxy.cpu().numpy().astype(int)  
    scores = results[0].boxes.conf.cpu().numpy()  
    class_ids = results[0].boxes.cls.cpu().numpy().astype(int)  
    for i, class_id in enumerate(class_ids):
        character = class_map[class_id]
        detectii_dict[character].append(detections[i])  
        scoruri_dict[character].append(scores[i])  
        file_names_dict[character].append(img_path.name)  
```

```python

for image_file in test_images_dir.glob("*.jpg"):  
    recunoastere_personaje(image_file, model)


for character in detectii_dict:
    if detectii_dict[character]: 
        np.save(f"./351_Pirvulescu_Daria/task2_bonus/detections_{character}.npy", np.array(detectii_dict[character]))
        np.save(f"./351_Pirvulescu_Daria/task2_bonus/scores_{character}.npy", np.array(scoruri_dict[character]))
        np.save(f"./351_Pirvulescu_Daria/task2_bonus/file_names_{character}.npy", np.array(file_names_dict[character]))
```
