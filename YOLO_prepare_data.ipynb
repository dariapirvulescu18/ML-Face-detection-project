```python
import os
from shutil import copy2
```

```python
def scrie_img_pt_YOLO(file_path_din_care_copiez, pers,output):
    files=os.listdir(file_path_din_care_copiez)
    for file in files:
        full_file_path = os.path.join(file_path_din_care_copiez, file)
        name, ext = os.path.splitext(file)
        new_file_name = f"{pers}_{name}{ext}"
        new_file_path = os.path.join(output, new_file_name)
        copy2(full_file_path, new_file_path)
```

```python
latime_img = 480
inaltime_img = 360
```

```python
def scrie_labels_pt_YOLO_task1(input_file, output_dir, word):
    os.makedirs(output_dir, exist_ok=True)
    prev_img = None
    new_filename = None
    with open(input_file, 'r') as infile:
        for line in infile:
            lista = line.split()
            img_name = lista[0]
            xmin= int(lista[1])
            ymin= int(lista[2])
            xmax= int(lista[3])
            ymax= int(lista[4])
            x_center = ((xmin + xmax) / 2) / latime_img
            y_center = ((ymin + ymax) / 2) / inaltime_img
            width = (xmax - xmin) / latime_img
            height = (ymax - ymin) / inaltime_img
            if prev_img != img_name:
                new_filename = os.path.join(output_dir, f"{word}_{img_name[:-4]}.txt")
                prev_img = img_name
    
            with open(new_filename, 'a') as outfile:
                outfile.write(f"0 {x_center} {y_center} {width} {height}\n")
```

```python
os.makedirs('./dataset_toate_fetele/dataset/train/labels/',exist_ok=True)
os.makedirs('./dataset_toate_fetele/dataset/val/labels/',exist_ok=True)


scrie_labels_pt_YOLO_task1('./antrenare/dad_annotations.txt','./dataset_toate_fetele/dataset/train/labels','dad')
scrie_labels_pt_YOLO_task1('./antrenare/deedee_annotations.txt','./dataset_toate_fetele/dataset/train/labels','deedee')
scrie_labels_pt_YOLO_task1('./antrenare/dexter_annotations.txt','./dataset_toate_fetele/dataset/train/labels','dexter')
scrie_labels_pt_YOLO_task1('./antrenare/mom_annotations.txt','./dataset_toate_fetele/dataset/train/labels','mom')
```

```python
def scrie_labels_pt_YOLO_task2(input_file, output_dir, word):
    os.makedirs(output_dir, exist_ok=True)
    dictionar_pers = {"dad":0, "deedee":1, "dexter":2, "mom":3}
    prev_img = None
    new_filename = None
    with open(input_file, 'r') as infile:
        for line in infile:
            lista = line.split()
            img_name = lista[0]
            # if img_name=='021.jpg':
            #     return 
            xmin= int(lista[1])
            ymin= int(lista[2])
            xmax= int(lista[3])
            ymax= int(lista[4])
            personaj = lista[5]
            if personaj== "unknown":
                continue
            x_center = ((xmin + xmax) / 2) / latime_img
            y_center = ((ymin + ymax) / 2) / inaltime_img
            width = (xmax - xmin) / latime_img
            height = (ymax - ymin) / inaltime_img
            if prev_img != img_name:
                new_filename = os.path.join(output_dir, f"{word}_{img_name[:-4]}.txt")
                prev_img = img_name
            clasa = dictionar_pers[personaj]
            with open(new_filename, 'a') as outfile:
                outfile.write(f"{clasa} {x_center} {y_center} {width} {height}\n")

       
```

```python
os.makedirs('./dataset_personaje/dataset/train/labels/',exist_ok=True)
os.makedirs('./dataset_personaje/dataset/val/labels/',exist_ok=True)

scrie_labels_pt_YOLO_task2('./antrenare/dad_annotations.txt','./dataset_personaje/dataset/train/labels','dad')
scrie_labels_pt_YOLO_task2('./antrenare/deedee_annotations.txt','./dataset_personaje/dataset/train/labels','deedee')
scrie_labels_pt_YOLO_task2('./antrenare/dexter_annotations.txt','./dataset_personaje/dataset/train/labels','dexter')
scrie_labels_pt_YOLO_task2('./antrenare/mom_annotations.txt','./dataset_personaje/dataset/train/labels','mom')

```
